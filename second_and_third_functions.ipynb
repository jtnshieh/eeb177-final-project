```python
def make_dolphin_dictionary_with_mean_age(filename):
    with open(filename, 'r') as ff:
        dolphin_recs = ff.readlines()
    from collections import defaultdict
    species_ranges = defaultdict(list)
    for line in dolphin_recs:
        species = line.split('","')[0]
        maxage = line.split('","')[2]
        minage = line.split('","')[3]
        mean_age = (float(minage) + float(maxage)) / 2
        species_ranges[species].append(mean_age)
    return species_ranges

make_dolphin_dictionary_with_mean_age("dolphin_age_data.csv")
```

```python
def make_list_of_oldest_and_youngest_for_each_dolphin_species(filename):
    result = make_dolphin_dictionary_with_mean_age(filename)
    dolphin_list = []
    for key in result.keys():
        ages = result[key]
        minage = min(ages)
        maxage = max(ages)
        dolphin_list.append(key + ", " + str(minage) + ", " + str(maxage))
    return dolphin_list

make_list_of_oldest_and_youngest_for_each_dolphin_species("dolphin_age_data.csv")
```
