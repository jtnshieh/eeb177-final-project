```python
def make_cetacea_dictionary_with_mean_age(filename):
    from collections import defaultdict
    genera = open(filename, 'r')
    genera.readline()
    species_ranges = defaultdict(list)
    for line in genera:
        species = line.split('","')[0]
        maxage = line.split('","')[2]
        minage = line.split('","')[3]
        mean_age = (float(minage) + float(maxage))/2
        species_ranges[species].append(mean_age)
    genera.close()
    return species_ranges

make_cetacea_dictionary_with_mean_age("reformatted_cetacea_pbdb_data.csv")
```

```python
def make_list_of_oldest_and_youngest_for_each_cetacea_species(filename, new_filename):
    output_file = open(new_filename, "w")
    result = make_cetacea_dictionary_with_mean_age(filename)
    for key in result.keys():
        ages = result[key]
        minage = min(ages)
        maxage = max(ages)
        genus = key.split(" ")[0].strip('"') 
        output_file.write(genus + "," + key.strip('"') + "," + str(minage) + "," + str(maxage) + "\n")
    output_file.close()
    return output_file

make_list_of_oldest_and_youngest_for_each_cetacea_species("reformatted_cetacea_pbdb_data.csv","data_for_graphing")
```
